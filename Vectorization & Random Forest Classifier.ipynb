```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DownloadS/Data.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
train = df[df['Date']<'2015']
test = df[df['Date']>'2015']
```

```python
train.tail()
```

```python
test.head()
```

```python
data = train.iloc[:,2:27]
```

```python
data.replace("[^a-zA-Z0-9\s]", " ", regex=True, inplace=True)
```

```python
data.head()
```

```python
data.columns=[str(i) for i in range(25)]
```

```python
data.head()
```

```python
for  i in range(25):
    data.iloc[i] = data.iloc[i].str.lower()
```

```python
data.head()
```

```python
" ".join(str(x) for x in data.iloc[0,:])
```

```python
headlines = []
for row in data.index:
    headlines.append(" ".join(str(x) for x in data.iloc[row,:]))
```

```python
headlines[9]
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
traindataset.toarray().shape
```

```python
countvector.get_feature_names_out()
```

```python
headlines[1]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(traindataset, train['Label'])
```

```python
test.head()
```

```python
testdf = test.iloc[:,2:27]
```

```python
testdf.head()
```

```python
testdf.replace("[^a-zA-Z0-9\s]", "", regex=True, inplace=True)
```

```python
testdf.index=list(range(len(testdf)))
```

```python
testdf.columns=list(range(25))
```

```python
for i in testdf.columns:
    testdf[i]=testdf[i].str.lower()
```

```python
testdf.head()
```

```python
testheadlines = []
for row in testdf.index:
    testheadlines.append(" ".join(str(x) for x in testdf.iloc[row,:]))
```

```python
testheadlines[9]
```

```python
testdata = countvector.transform(testheadlines)
```

```python
print("Vocabulary:",countvector.get_feature_names_out())
```

```python
predictions = randomclassifier.predict(testdata)
```
