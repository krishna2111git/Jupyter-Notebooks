```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_with_discounts.csv')
df
```

```python
model_net_sales = df.groupby(['BU', 'Brand', 'Model'])['Net Sales Value'].sum().astype(int).to_frame()
model_net_sales
#Revenue from each model of each brand
```

```python
mbl_net_sls = model_net_sales.loc['Mobiles','Net Sales Value'].sum()
lfs_net_sls = model_net_sales.loc['Lifestyle','Net Sales Value'].sum()
fmcg_net_sls = model_net_sales.loc['FMCG','Net Sales Value'].sum()
```

```python
BU_net_sls = model_net_sales.index.get_level_values('BU').map({'Mobiles': mbl_net_sls , 'Lifestyle': lfs_net_sls, 'FMCG':fmcg_net_sls})
```

```python
model_net_sales['BU Net Sales'] = BU_net_sls
```

```python
model_net_sales
```

```python
df.groupby('Brand')['Net Sales Value'].sum().astype(int)
```

```python
brand_sales = model_net_sales.groupby('Brand')['Net Sales Value'].sum().to_list()
brand_sales
```

```python
brands = set(model_net_sales.index.get_level_values('Brand'))
brands
```

```python
map_dict = {Brand: sales for Brand, sales in zip(brands, brand_sales)}
```

```python
model_net_sales['Brand Net Sales'] = model_net_sales.index.get_level_values('Brand').map(map_dict)
```

```python
model_net_sales 
```

```python
net_sales = model_net_sales.groupby(['BU Net Sales', 'Brand Net Sales'])['Net Sales Value'].apply(lambda x:x).to_frame()
```
