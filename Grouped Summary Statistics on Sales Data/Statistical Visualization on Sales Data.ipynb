```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_with_discounts.csv')
df
```

```python
df.describe()
```

```python
total_volume_each_day_each_bu = df.groupby(['Date', 'BU'])['Volume'].sum().to_frame()
total_volume_each_day_each_bu
```

```python
fmcg_volume_count = total_volume_each_day_each_bu.xs('FMCG', level='BU').value_counts().to_frame().reset_index()
fmcg_volume_count
```
